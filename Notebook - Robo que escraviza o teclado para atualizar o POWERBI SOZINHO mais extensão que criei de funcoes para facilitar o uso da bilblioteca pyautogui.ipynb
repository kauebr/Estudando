```python
import pyautogui
import time
from PIL import ImageGrab
from pyautogui import ImageNotFoundException
```

```python
# VPN tem que estar conectada 
def inicio():
    # 1 - Minimiza todas as janelas
    pyautogui.hotkey('win', 'd')

    # 2 - Espera 3 segundos
    time.sleep(3)

    # 3 - Mouse canto superior direito
    largura, _ = pyautogui.size()
    pyautogui.moveTo(largura - 5, 5, duration=0.5)

def executar(caminho):
    """
    Abre um aplicativo usando Win + R.

    :param caminho: Caminho ou comando a ser digitado no Executar (Win + R)
    :param tempo_espera: Tempo (em segundos) para aguardar após abrir
    """
    pyautogui.hotkey('win', 'r')
    time.sleep(1)

    pyautogui.write(caminho, interval=0.01)
    pyautogui.press('enter')

def capturar_posicao(
        tempo_espera):
    """
    Aguarda um tempo e captura a posição atual do mouse.

    :param tempo_espera: Tempo (em segundos) para aguardar antes de capturar
    """
    time.sleep(tempo_espera)
    x, y = pyautogui.position()
    print(f'pyautogui.moveTo({x}, {y})')

def capturar_imagem(nome_arquivo):

    """
    Captura uma screenshot da tela inteira e salva em um arquivo.

    :param nome_arquivo: Nome do arquivo onde a imagem será salva
    """
    time.sleep(5)
    pyautogui.hotkey("win", "shift", "s")
    time.sleep(5)
    # espera até o usuário capturar a tela
    while ImageGrab.grabclipboard() is None:
        time.sleep(0.5)
    ImageGrab.grabclipboard().save(nome_arquivo)

def esperar_imagem_carregar_e_clicar(
    nome_arquivo,
    timeout=600,
    confidence=0.8
):
    """
    Espera até que uma imagem apareça na tela e clica no centro.
    
    :param nome_arquivo: caminho da imagem
    :param timeout: tempo máximo de espera (segundos)
    :param confidence: confiança da detecção
    :return: True se clicou, False se deu timeout
    """
    inicio = time.time()

    while True:
        try:
            pos = pyautogui.locateOnScreen(nome_arquivo, confidence=confidence)

            if pos:
                x, y = pyautogui.center(pos)
                pyautogui.moveTo(x, y, duration=0.3)
                pyautogui.click()
                return True

        except ImageNotFoundException:
            pass  # imagem ainda não apareceu, continua tentando

        if time.time() - inicio > timeout:
            print(f"⚠️ Timeout: imagem '{nome_arquivo}' não apareceu.")
            return False
        time.sleep(0.5)

def esperar_ate_frame_sumir(
    imagem,
    timeout=10000,
    intervalo=10,
    confidence=0.8
):
    """
    Espera até a imagem SUMIR da tela.

    :param imagem: caminho da imagem
    :param timeout: tempo máximo de espera (segundos)
    :param intervalo: tempo entre verificações
    :param confidence: confiança da detecção
    """
    time.sleep(25)
    inicio = time.time()

    while True:
        try:
            pyautogui.locateOnScreen(imagem, confidence=confidence)
            # Se achou, continua esperando
            time.sleep(intervalo)

            if time.time() - inicio > timeout:
                print("⚠️ Timeout atingido, imagem ainda visível.")
                return False

        except ImageNotFoundException:
            # NÃO achou → imagem sumiu → sucesso
            return True
        
def email_logado():

    """Captura em qual conta estou logado no RADAR FDL"""
    time.sleep(5)
    try:
        pos = pyautogui.locateOnScreen('conta_logada.png', confidence=0.8)
        return "fdl"
    except ImageNotFoundException:
        return "k8"
```

```python
import time
import ctypes
import threading
import sys
import pyautogui
import tkinter as tk    

# ===============================
# BLOQUEIO / DESBLOQUEIO INPUT
# ===============================
def bloquear_input():
    ctypes.windll.user32.BlockInput(True)

def liberar_input():
    ctypes.windll.user32.BlockInput(False)

# ===============================
# POPUP STATUS (inferior direito)
# ===============================
class PopupStatus:
    def __init__(self, tempo_leitura=10):
        self.tempo_leitura = tempo_leitura

        self.root = tk.Tk()
        self.root.title("RADAR")
        self.root.overrideredirect(True)
        self.root.attributes("-topmost", True)

        largura = 320
        altura = 160
        margem = 40

        tela_larg, tela_alt = pyautogui.size()
        x = tela_larg - largura - margem
        y = tela_alt - altura - margem

        self.root.geometry(f"{largura}x{altura}+{x}+{y}")

        self.status = tk.StringVar()

        self.label = tk.Label(
            self.root,
            textvariable=self.status,
            font=("Segoe UI", 10),
            wraplength=300,
            justify="left"
        )
        self.label.pack(padx=10, pady=20)

        self.root.update()

    def mostrar(self, mensagem):
        self.status.set(mensagem)
        self.root.update()
        time.sleep(self.tempo_leitura)

    def fechar(self):
        self.root.destroy()

# ===============================
# EXECUÇÃO RADAR
# ===============================
def executar_radar():
    inicio = time.time()

    popup.mostrar("🚀 Iniciando atualização do RADAR Executivo")
    bloquear_input()

    popup.mostrar("🔒 Teclado e mouse bloqueados\nAguarde...")

    executar(
        r"C:\Users\Kaue Mandarino\OneDrive - FABRICA DO LIVRO GRAFICA E EDITORA LTDA\Fabrica_Pasta_Geral\Setor de Dados\Escriba FDL\RADAR FDL 2.0\Radar FDL 2.pbix"
    )

    time.sleep(60)

    popup.mostrar("📂 Radar aberto")
    esperar_imagem_carregar_e_clicar("screenshot_radar_aberto.png", confidence=0.95)

    popup.mostrar("🔄 Atualizando dados...")
    esperar_ate_frame_sumir("frame_atualizar.png")

    popup.mostrar("📤 Publicando no Power BI...")
    time.sleep(10)
    esperar_imagem_carregar_e_clicar("arquivo.png", confidence=0.9)
    time.sleep(10)
    esperar_imagem_carregar_e_clicar("publicar.png", confidence=0.9)
    time.sleep(10)
    esperar_imagem_carregar_e_clicar("publicar_no_PBI.png", confidence=0.9)
    time.sleep(10)
    esperar_imagem_carregar_e_clicar("scratch.png", confidence=0.9)
    time.sleep(10)
    esperar_imagem_carregar_e_clicar("selecionar.png", confidence=0.9)
    time.sleep(10)
    esperar_imagem_carregar_e_clicar("substituir.png", confidence=0.9)

    fim = time.time()
    tempo = round(fim - inicio, 1)

    liberar_input()

    popup.mostrar(
        f"✅ Atualização concluída!\n⏱️ Tempo total: {tempo}s\n🔓 Sistema liberado"
    )
    time.sleep(10)
    popup.fechar()


import tkinter as tk
import time
import sys

def solicitar_autorizacao_atualizacao():
    """
    Abre um popup perguntando se o usuário deseja:
    - Autorizar agora
    - Esperar mais 30 minutos
    - Cancelar
    Retorna: "autorizar", "esperar", "cancelar"
    """

    escolha = {"acao": None}

    root = tk.Tk()
    root.title("Atualização do RADAR")
    root.attributes("-topmost", True)
    root.resizable(False, False)

    largura = 360
    altura = 180
    margem = 60

    tela_larg = root.winfo_screenwidth()
    tela_alt = root.winfo_screenheight()

    x = tela_larg - largura - margem
    y = tela_alt - altura - margem

    root.geometry(f"{largura}x{altura}+{x}+{y}")

    msg = tk.Label(
        root,
        text="⏰ Hora de atualizar o RADAR Executivo.\n\nDeseja autorizar a atualização agora?",
        font=("Segoe UI", 10),
        wraplength=330,
        justify="left"
    )
    msg.pack(pady=20)

    frame_botoes = tk.Frame(root)
    frame_botoes.pack(pady=10)

    def autorizar():
        escolha["acao"] = "autorizar"
        root.destroy()

    def esperar():
        escolha["acao"] = "esperar"
        root.destroy()

    def cancelar():
        escolha["acao"] = "cancelar"
        root.destroy()

    tk.Button(frame_botoes, text="✅ Autorizar", width=10, command=autorizar).grid(row=0, column=0, padx=5)
    tk.Button(frame_botoes, text="⏳ Esperar 30 min", width=12, command=esperar).grid(row=0, column=1, padx=5)
    tk.Button(frame_botoes, text="❌ Cancelar", width=10, command=cancelar).grid(row=0, column=2, padx=5)

    root.mainloop()
    return escolha["acao"]


# ===============================
# EXEMPLO DE USO
# ===============================
acao = solicitar_autorizacao_atualizacao()

if acao == "autorizar":
    print("Atualização autorizada. Iniciando...")
elif acao == "esperar":
    print("Aguardando 30 minutos...")
    time.sleep(1800)
    print("Retomando atualização...")
elif acao == "cancelar":
    print("Atualização cancelada pelo usuário.")
    sys.exit()

# ===============================
# START
# ===============================
popup = PopupStatus(tempo_leitura=10)

threading.Thread(target=executar_radar, daemon=True).start()

popup.root.mainloop()
```
